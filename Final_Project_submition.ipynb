# Dealing with missing value

```python
raw_data['Credit_History'].fillna(2, inplace= True)
test_data['Credit_History'].fillna(2, inplace=True)
```

```python
a = ['Gender', 'Married', 'Dependents', 'Self_Employed']
for i in range(len(a)):
    raw_data[f'{a[i]}'].fillna(method='ffill', inplace=True)
    test_data[f'{a[i]}'].fillna(method='ffill', inplace=True)
    
b = ['LoanAmount', 'Loan_Amount_Term']
for i in range(2):
    raw_data[f'{b[i]}'].fillna(raw_data[f'{b[i]}'].median(), inplace=True)
    test_data[f'{b[i]}'].fillna(test_data[f'{b[i]}'].median(), inplace=True)
raw_data.isnull().sum()
```

```python
c = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Credit_History', 
    'Property_Area', 'Loan_Status']
for i in range(len(c)):
    raw_data = pd.get_dummies(data=raw_data, columns=[f'{c[i]}'], drop_first=True)

d = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Credit_History', 
    'Property_Area']
for j in range(len(d)):
    test_data = pd.get_dummies(data=test_data, columns=[f'{d[j]}'], drop_first=True)
row_data = raw_data.shape[0]
col_data = raw_data.shape[1]
print('The train data after dealing with missing value and convert to categorical have {} row and {} columns'.format(row_data, col_data))
```

# Create 3 new columns are ''Total_Income", "Debt_To_Income" and "Total_Income_Ratio"

```python
raw_data['Total_Income'] = raw_data['ApplicantIncome'] + raw_data['CoapplicantIncome']
raw_data['Total_Income_Ratio'] = raw_data['Total_Income']/ raw_data['Loan_Amount_Term']
raw_data['Debt_To_Income'] = raw_data['Total_Income']/ raw_data['LoanAmount']

test_data['Total_Income'] = test_data['ApplicantIncome'] + test_data['CoapplicantIncome']
test_data['Total_Income_Ratio'] = test_data['Total_Income']/ test_data['Loan_Amount_Term']
test_data['Debt_To_Income'] = test_data['Total_Income']/ test_data['LoanAmount']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
rfc = RandomForestClassifier(n_estimators=200, max_depth=50)
dtc = DecisionTreeClassifier()
```

```python
X = raw_data.drop(columns=['Loan_Status_Y'])
y = raw_data['Loan_Status_Y']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 101)
```

# Oversampling data

```python
train_data = pd.concat([X_train, y_train], axis=1)
```

```python
over_data_Y = train_data[train_data['Loan_Status_Y']==1]
over_data_N = train_data[train_data['Loan_Status_Y']==0]
```

```python
oversampling_N = over_data_N.sample(replace=True, n = over_data_Y.shape[0])
Total_data = pd.concat([over_data_Y, oversampling_N], axis=0)
sns.countplot(data=Total_data, x='Loan_Status_Y')
```

```python
X_train_over = Total_data.drop(columns=['Loan_Status_Y'])
y_train_over = Total_data['Loan_Status_Y']
```

```python
rfc.fit(X_train_over, y_train_over)
dtc.fit(X_train_over, y_train_over)
```

```python
rfc_predict = rfc.predict(X_test)
dtc_predict = dtc.predict(X_test)
```

```python
print(accuracy_score(y_test, rfc_predict))
print(accuracy_score(y_test,dtc_predict))
```

```python
print(classification_report(y_test, rfc_predict))
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
param_Grid = {
    'n_estimators': [100,200,300,500],
    'max_depth': [50,80,100]
}
gridcv = GridSearchCV(rfc, param_grid=param_Grid, n_jobs=4, verbose=2)
gridcv.fit(X_train_over, y_train_over)
```
